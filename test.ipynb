```python
frame = cv2.imread("./images/base.png")
overlay = cv2.imread("./images/danger.png")
H, W = overlay.shape[:2]
frame = cv2.resize(frame, (W, H))

print(frame.shape)
print(overlay.shape)

s = cv2.addWeighted(
                    src1=frame, alpha=0.6, src2=overlay, beta=0.3, gamma=0
                )
import matplotlib.pyplot as plt
plt.imshow(s)
```
